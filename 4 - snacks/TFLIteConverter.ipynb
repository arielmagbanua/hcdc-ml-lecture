```python
import tensorflow as tf
from google.colab import files
```

```python
# load the model
model = tf.keras.models.load_model('snacks_classifier.keras')

# Convert the model.
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the model.
with open('snacks_classifier.tflite', 'wb') as f:
  f.write(tflite_model)

files.download('snacks_classifier.tflite')
```
